```python
import pandas as pd
import numpy as np

from pycaret.classification import *

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("df_t2dm.csv")
print(df.shape)
df.head()
```

```python
df_filtered = df.copy()

# Define your variables
target_variable = 'has_diabetes'
individual_level_vars = ['age_cate', 'female', 'education', 'wealth', 'notwork', 'smoking']
community_level_vars = ['com_poverty', 'com_illit', 'com_rural', 'com_uemp']
categorical_vars = ['age_cate', 'female', 'education', 'wealth', 'notwork', 'smoking']
```

```python
# Step 1: Balance the dataset to prevent India bias
max_samples_per_country = 40000  # max samples to keep per country

df_balanced = df_filtered.groupby('countryname').apply(
    lambda x: x.sample(n=min(len(x), max_samples_per_country), random_state=123)
).reset_index(drop=True)

print(f"Original dataset size: {len(df_filtered)}")
print(f"Balanced dataset size: {len(df_balanced)}")

print("\nCountry distribution after balancing:")
print(df_balanced['countryname'].value_counts())

# Keep only South Africa
df_balanced = df_balanced[df_balanced['countryname'] == 'South Africa']

print("\nAfter filtering South Africa:")
print(df_balanced['countryname'].value_counts())
```

```python
# Step 2: Identify continuous variables for normalization
continuous_individual = [v for v in individual_level_vars
                         if v not in categorical_vars and v != target_variable]
continuous_community = [v for v in community_level_vars
                        if v not in categorical_vars]

all_continuous = continuous_individual + continuous_community

print("Individual continuous variables:", continuous_individual)
print("Community continuous variables:", continuous_community)
print("\nAll continuous variables to normalize:", all_continuous)

# Step 3: Normalize continuous features within countries
df_normalized = df_balanced.copy()

for var in all_continuous:
    if var in df_normalized.columns:
        df_normalized[f'{var}_norm'] = df_normalized.groupby('countryname')[var].transform(
            lambda x: (x - x.mean()) / x.std() if x.std() > 0 else 0
        )
        print(f"Normalized: {var}")
```

```python
# Step 4: Add country as a feature
df_normalized['country_encoded'] = pd.Categorical(df_normalized['countryname']).codes

# Step 5: Prepare datasets for both models

# Model 1: Individual-level variables only
individual_categorical = [v for v in individual_level_vars if v in categorical_vars]
individual_continuous_norm = [f'{v}_norm' for v in continuous_individual if v in df_normalized.columns]

individual_features = individual_categorical + individual_continuous_norm + ['country_encoded']

df_individual_model = df_normalized[individual_features + [target_variable]].copy()

# Model 2: All variables (individual + community)
community_continuous_norm = [f'{v}_norm' for v in continuous_community if v in df_normalized.columns]
all_features = individual_features + community_continuous_norm

df_all_model = df_normalized[all_features + [target_variable]].copy()

print(f"Individual model features ({len(individual_features)}): {individual_features}")
print(f"All-variables model features ({len(all_features)}): {all_features}")
print("\nShapes:")
print("df_individual_model:", df_individual_model.shape)
print("df_all_model:", df_all_model.shape)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("\n" + "="*50)
print("MODEL 1: INDIVIDUAL-LEVEL VARIABLES ONLY (Random Forest, NO TUNING)")
print("="*50)

# 1) Setup for individual-level model
clf_individual = setup(
    data=df_individual_model,
    target=target_variable,
    fix_imbalance=True,
    session_id=123,
    train_size=0.8,
    verbose=False
)

# 2) Compare all models (output is printed to the screen)
compare_models(
    fold=5,
    sort="Recall",
)

# 3) Creating a Random Forest model without tuning
rf_individual = create_model('rf')

# 4) Finalise RF on the entire train data
final_individual = finalize_model(rf_individual)

# 5) Getting predictions on the test set with the final RF model
predict_individual = predict_model(final_individual)

print("\nFinal individual-level model (Random Forest, no tuning):")
print(final_individual)

# 6) Final report on the test set
y_true = predict_individual[target_variable]
y_pred = predict_individual['prediction_label']

print("\nClassification Report (Test Set, RF no tuning):")
print(classification_report(y_true, y_pred))

print("\nConfusion Matrix (Test Set, RF no tuning):")
print(confusion_matrix(y_true, y_pred))
```

```python
print("\n" + "="*50)
print("MODEL 2: ALL VARIABLES (INDIVIDUAL + COMMUNITY)")
print("="*50)

#1) Hybrid data setup
clf_all = setup(
    data=df_all_model,
    target=target_variable,
    fix_imbalance=True,
    session_id=123,
    train_size=0.8,
    verbose=False
)

# 2) Show performance of all models 
compare_models(fold=5, sort="Recall")

# 3) Building a Random Forest Model
rf_all = create_model('rf')

# 4) RF finalisation on the entire train data
final_all = finalize_model(rf_all)

# 5) Prediction on the test set
predict_all = predict_model(final_all)

print("\nFinal model (Model 2 - RF no tuning):")
print(final_all)

# 6) Performance report
from sklearn.metrics import classification_report, confusion_matrix

y_true = predict_all[target_variable]
y_pred = predict_all['prediction_label']

print("\nClassification Report (Model 2):")
print(classification_report(y_true, y_pred))

print("\nConfusion Matrix (Model 2):")
print(confusion_matrix(y_true, y_pred))
```
